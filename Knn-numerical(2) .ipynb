```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error, r2_score
```

```python
train=pd.read_csv(r"C:\Users\Mai\Downloads\1553768847-housing.csv")
train.head(10)
```

```python
x= train.drop('median_house_value', axis=1)  
y= train['median_house_value'] 
```

```python
x.isnull().sum()
```

```python
x['total_bedrooms'] = x['total_bedrooms'].fillna(x['total_bedrooms'].mean())
```

```python
x.isnull().sum()
```

```python
x = pd.get_dummies(x, columns=['ocean_proximity'])
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaling = scaler.fit_transform(x_train)
x_test_scaling = scaler.transform(x_test)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(x_train_scaling, y_train)
```

```python
y_pred= knn_model.predict(x_test_scaling)
```

```python
y_pred
```

```python
y_pred = pd.DataFrame(y_pred)
y_pred
```

```python
y_test
```

```python
y_test= pd.DataFrame(y_test)
y_test
```

```python
y_test = pd.DataFrame(y_test).reset_index(drop=True)
y_test
```

```python
from sklearn.metrics import mean_squared_error, root_mean_squared_error

mse = mean_squared_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
```

```python
r_squared = r2_score(y_test, y_pred)
print(f"R-squared score: {r_squared}")
```
